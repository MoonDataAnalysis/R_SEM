# 1.1、相关关系/共变关系



两种计算相关关系的指标：

covariance：反应两变量共变关系，没标准化，不同covariance不能相互比较



correlation：反应两变量共变关系，标准化，不同correlation可以相互比较



# 1.2、回归关系





反映自变量对因变量的预测关系

# 1.3、路径模型



路径模型可能包含相关关系，也可能包含回归关系，也可能兼而有之

# 1.4、因子模型



多个显变量测量一个潜变量

# 1.5、结构方程模型



结构方程模型即带有潜变量的路径模型

# 2、用结构方程模型做研究，发表期刊文章的流程




#### 2.1、定义理论框架：有理有据提出理论框架

1. 所谓有“理”：理论框架不是凭空而来，而是有所来源

2. 所谓有“据”：理论框架有实证研究的支持

3. 所谓理论框架：是你对现象的预期，包含若干核心假设，假设要提清楚

例如：葛枭语, & 侯玉波. (2021). 君子不忧不惧: 君子人格与心理健康——自我控制与真实性的链式中介. 心理学报, 53(4), 374.



#### 2.2、按照理论框架定义结构方程模型



#### 2.3、明确需要测量哪些变量：测量工具要有出处有依据

1. 所谓有“出处”：测量工具有明确来源

2. 所谓有“依据”：用在你的场景恰当

#### 2.4、收集数据： 数据量要充足、采集数据有代表性

1. 所谓数据量充足：你的模型需要多少样本量？

2. 你的数据是否有代表性：尽量随机取样

#### 2.5、模型估计

```r
# 在lavaan中定义模型
MOD <- "

   # 定义结构模型
    自我控制 ~ a1 * 君子人格
    真实性 ~ a2 * 君子人格 + b2 * 自我控制
    心理健康 ~  d * 君子人格 + c1 * 自我控制 + c2 * 真实性
    
    # 定义各中介效应
    pri_indeff_m1 := a1*c1
    pri_indeff_m2 := a2*c2
    sec_indeff_m1 := a1*b2
    sec_indeff_m2 := b2*c2
    ind_eff := a1 * b2 * c2

"
# 拟合模型
MOD_fit <- sem(MOD,BASE)

# 提取统计模型
summary(MOD_fit,
        fit.measures = TRUE, 
        standardized = TRUE,
        modindices = TRUE, 
        rsq = TRUE)
```
