# 如何计算统计功效或样本量呢？

### 方案二： RMSEA法—— RMSEA-based power caculation 


### 故事需要从检验SEM模型拟合说起：

（建议大家去看一下第三课“结构方程模型的估计流程和估计方法”）

为检验SEM模型是否是实际数据相符，提出H0假设，实际数据和模型大体相同；H1假设，实际数据与模型有显著差异



H0代表模型与数据完全相符，其expected χ2可以表达为：

E(χ2) = E(sampling error) = df

H1代表模型与数据不完全相符，其的expected χ2可以表达为：

E(χ2) = E(sampling error) + E(misspecification error) = df + λ




如果H1的统计值（E(χ2)）落在H0的拒绝阈外，则接受H0，说明模型与数据相符，反之则说明模型与实际数据有出入。

### 这是在检验模型拟合是否足够差。
### 那么换一种角度，我们检验拟合是否足够好，借助拟合这个“桥梁”，把统计功效和样本量联系起来

### 介绍今天的主角：RMSEA
RMSEA衡量每个df承受的misspecification，RMSEA越小，则模型与数据越贴合。 

是衡量SEM模型拟合最重要、最常用的指标之一。



RMSEA好坏评判有明确的范围：



不同的RMSEA值可以帮助确定不同的χ2分布，基于此，MacCallum et al. (1996) 提出用RMSEA确定power或样本量的思想

### 基于拟合做power analysis的核心思想

如果你确定最差的RMSEA，你的模型最少多少样本量，才能尽可能不超过这个RMSEA？

举例：

我们一般认为RMSEA=0.10，是模型是底线了

H0：模型拟合不好， RMSEA>0.10

H1：模型拟合较好， RMSEA<0.10

假如我们通过预实验也好，基于前人文献也好，确定模型RMSEA=0.05, 我们要与0.10比较，判断是多大程度能拒绝H0（也就是拒绝bad fit）

### RMSEA法的优点
1. 极其简单
2. 越来越流行

### RMSEA法的缺点
1. 只适用于正态数据
2. 不适用于multilevel SEM
