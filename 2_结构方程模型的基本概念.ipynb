## 思考：相关等于因果吗？

相关的计算只能告诉我们两个变量之间的共变关系。
要满足因果探讨，有3个标准：
1. 两个变量统计上存在关联（the association rule）
2. 原因先于结果，并明确的方向性（causal order）
3. 移除其他变量，因与果之间的关联不消失（the non-artificiality rule）

# 6.  Data distribution normality

我们需要明确多大程度所得参数是可以接受的，为此 SEM 依赖一个重要假设，即：正态性假设

如果模型中只有一个变量，那么做SEM 需要这个变量是正态的

如果模型中有多个变量，那么做SEM 需要多个变量是否正态！

###    
###  Why？

SEM衡量所得参数是否显著大于零，当（B-0）/SE 落在置信水平（95%）之外时，我们拒绝0假设。这里使正态分布的概率密度函数作为判断的依据，因此也要求该变量样本数据是正态分布的。

###    
### 如何验证正态性假设？

#### 单变量时:

 峰度（kurtosis）：太尖？太平？
 
 偏度（skewness）：太歪？
