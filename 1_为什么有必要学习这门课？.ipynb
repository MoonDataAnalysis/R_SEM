# 1. 为什么要学习结构方程模型？

结构方程模型现已是社科领域最重要的统计技术之一，

对于科学研究和学术论文投稿非常重要！

##### （1）结构方程模型在行为科学（心理学、管理学、教育学等等）中应用：

例如：自我效能与组织考虑的关系：中介作用



文献出处：

Weston, R., & Gore, P.A. (2006). A Brief Guide to Structural Equation Modeling. The Counseling Psychologist, 34, 719 - 751.

##### （2）结构方程模型在医学（尤其是精神医学应用较多）中的应用：

例如：精神分裂患者，阳性病症（幻觉、思维混乱、异想）、精神分裂持续时间对注意力和偏侧之间的关系

两个因变量：通过患者对通过耳机说话的辅音和元音混合的反应来测量二元听力。在一种情况下，患者被告知用哪只耳朵听(注意)，而在另一种情况下，他们不被告知(侧耳)。



文献出处：

Beran, T. N., & Violato, C. (2010). Structural equation modeling in medical research: a primer. BMC research notes, 3(1), 1-10.

##### (3) 结构方程模型在商学中的应用：

例如：主管多元文化能力与咨询自我效能的关系：联合工作监督的中介作用



文献出处：

Vavere, E. (2019). Investigating the usefulness of structural equation modelling for service quality in a SaaS delivery model.

##### （4）结构方程模型在体育领域应用：

例如： 运动员与教练的关系质量对成就感知、心理与情绪消耗、运动员野心下降（sport devaluation）的影响：运动员掌握趋近目标与掌握回避目标的中介作用



##### 还有许多其他领域......





# 2. 为什么借助 R 语言学习结构方程模型？优势是？

##### 为什么不用AMOS、SPSS、Mplus？



1、R语言与上述软件能实现的模型呈包含关系，上述软件能实现的，R一定也能实现，R能实现的，上述软件不一定能实现。

2、前沿结构方程模型应用都是用R实现的。

3、R本身是非常好用的工具。学会R语言，你将再不愿意重新使用AMOS/SPSS/Mplus...
