# 如果使用量表工具测量，需要提供量表的信度和效度指标

### 那么什么是信度，什么是效度？


信度反应测量工具的可靠性或一致性，测量工具内部各条目（题目）不能一个问A，一个问B；不同时间点，测量结果一致，不能一个时间点A，一个时间点B

效度反应测量能真实测量目标内容

效度高，信度一定高；信度高，效度不一定高。

# 效度测量

### 1、结构效度

主要考察量表的内部结构（主要是维度划分）是否与实际数据一致。例如，工作倦怠量表（Maslach Burnout Inventory,MBI）从三个维度衡量工作倦怠（情绪耗竭、懈怠、成就感低），你用该量表收了数据。如果样本缺少可以从这三个维度衡量，那么结构效度高，如果不能，结构效度低。

验证结构效度质量最常用的方法是验证性因素分析法：

```r
juand <- read.csv("C:/Users/77387/Desktop/Data_analysis_courses/R语言与结构方程模型/data/juandai.csv")

head(juand)
```

```r
Mod <- "

情绪耗竭 =~ 倦怠2 + 倦怠5 + 倦怠8 + 倦怠12 

懈怠 =~ 倦怠3 + 倦怠6 + 倦怠9 + 倦怠10 +  倦怠13

低成就感 =~ 倦怠1 + 倦怠4 + 倦怠7 + 倦怠11 +  倦怠14 + 倦怠15 + 倦怠16
"
```

```r
library(lavaan)
```

```r
Mod_fit <- sem(Mod , data = juand)
```

```r
fitMeasures(Mod_fit,c("chisq","df","pvalue","cfi","tli","rmsea"))
```

#### 验证性因素分析模型通过，则说明该量表应用在你的样本集上结构效度没有问题


### 2、内容效度

内容效度衡量所要测量的内容是否与要测量的内容一致，最长用的方法是专家评估法。专家根据自己的经验对内容效度做出评价。

### 3、区分效度（discriminant validity）

区分效度即该量表测量结果与其他高相关的变量（量表）能明显区分开

使用 semTools 函数库中的 discriminantValidity() 可以检验区分效度

```r
library(semTools)
HS.model <- ' visual  =~ x1 + x2 + x3
              textual =~ x4 + x5 + x6
              speed   =~ x7 + x8 + x9 '

fit <- cfa(HS.model, data = HolzingerSwineford1939)
discriminantValidity(fit)
discriminantValidity(fit, merge = TRUE)
```
