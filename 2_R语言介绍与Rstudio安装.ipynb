# 1、R 语言是什么？与Rstudio的关系？

R 语言是一种强力的统计绘图语言，在两方面及其突出：统计分析、数据可视化

R 语言的前身是 S 语言，出自赫赫有名的贝尔实验室（除了S语言之外，Unix、C语言、C++都是贝尔实验室研究人员首先开发的）



R 是一个开源平台，代码全部公开，对任何人免费。正是由于他开源的特性吸引了多领域的研究者不停地开发各种程序包工具（package，或“函数库”）

# 2、使用Rstudio

#### R 语言的“四大框”


#### R 语言的工具栏

1. File ——> New File :各种类型的脚本文件，通常只需用“R script”

2. Plot ——> Save as Image:把图保存为图片（.png/.tiff/.jpg等）

3. Session ——> Load Workspace/Save Workspace as/clear Workspace：把某 .Rdata中的变量导入工作区/保存工作区变量/清除工作区变量

4. Debug：用来找BUG，暂时用不到
