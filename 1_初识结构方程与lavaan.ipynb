# 1. 为什么要学习结构方程模型？或者说结构方程模型的应用场景是什么？


结构方程模型的两大优势：第一在于描绘和验证模型假设； 第二在于可以探讨潜变量。这两者对与很多社科领域的理论问题至关重要！


1. 第一个例子：什么是智力？
“智力”是一个完全抽象的东西，理论家基于某种方式提出一种理论，认为什么是智力。结构方程模型可以帮助你验证你的理论是否合理。



2. 第二个例子：流体智力影响口语能力？
理论家基于前人的研究或者观察推断流体智力能够影响口语能力。结构方程模型可以验证你的假设是否合理。

# 5. 什么是路径模型？
路径模型是比结构方程（或者潜变量建模）更大的范畴，泛指一切基于一定理论描绘变量关系。由于旨在回答变量之间影响的路径，所以称为路径模型。

在图论中路径模型之间的几何点称之为节点（nodes），链接节点的路径叫做边（edges）。
